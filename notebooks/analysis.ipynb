# Thematic Alignment Analysis
This notebook presents the alignment score analysis for the NLP project.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../results/alignment_scores.csv")
df.head()
```

```python
df.groupby("year")["alignment_score"].mean().plot()
plt.title("Alignment Over Time")
plt.xlabel("Year")
plt.ylabel("Mean alignment score")
plt.show()
```

```python
df["alignment_score"].hist()
plt.title("Alignment Score Distribution")
plt.xlabel("Alignment score")
plt.ylabel("Frequency")
plt.show()
```
